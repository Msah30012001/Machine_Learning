```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as mp
from sklearn import linear_model
```

```python
df = pd.read_csv('/content/medical.csv')
```

```python
df
```

```python
df.smoker[df.smoker == 'yes'] = 1
df.smoker[df.smoker == 'no'] = 0
```

```python
df
```

```python
%matplotlib inline 
mp.xlabel('age') 
mp.ylabel('smoker')
mp.scatter(df.age,df.smoker,color='red',marker='.')
```

```python
reg = linear_model.LinearRegression()
# fit method calculate m and c nothing but slope and y-intercept
reg.fit(df[['age']],df.smoker)
```

```python
reg.predict([[19]])
```
